### 1d(1)Images in each class are given in separate folders in the folder Sport Images. We have several hundreds of images for five classes of sport activities, namely baseball, basketball, fencing, tennis, and volleyball.

### 1d(2) Randomly select [0.7ni] images from each folder as your training set, [0.15ni] as validation set, and the rest as your test set, where ni is the number of images in folder i and /x/ is the ceiling of x

```python
import os
import random
import math
import shutil

# Define the source directory where the class folders are located
src_dir = "../Data/Sport_Images/"

# Define the destination directories for the training, validation, and test sets
train_dir = "../Data/train/"
val_dir = "../Data/val/"
test_dir = "../Data/test/"

# Define the class names
class_names = ["baseball", "basketball", "fencing", "tennis", "volleyball"]

# Define the proportions for the sets
train_prop = 0.7
val_prop = 0.15
test_prop = 0.15


train_X = []
val_X = []
test_X = []

train_Y = []
val_Y = []
test_Y = []
```

```python

for class_name in class_names:
    # Define the source directory for the current class folder
    class_dir = os.path.join(src_dir, class_name)
    
    # Get a list of all the image filenames in the class folder
    image_filenames = os.listdir(class_dir)
    
    # Get the number of images in the class folder
    num_images = len(os.listdir(class_dir))
    
    
    # Calculate the number of images for each set
    
    #train
    train_num = math.ceil(train_prop * num_images)
    train_subset = random.sample(image_filenames,k=train_num)
    train_X.extend([os.path.join(train_dir,class_name,str(i)) for i in train_subset])
    train_Y.extend([class_name.split('.')[-1]]*train_num)
    
    remaining_elements = [file for file in image_filenames if file not in train_subset]
    
    
    #val
    val_num = math.ceil(val_prop * num_images)
    val_subset = random.sample(remaining_elements,k=val_num)
    val_X.extend([os.path.join(val_dir,class_name,str(i)) for i in val_subset])
    val_Y.extend([class_name.split('.')[-1]]*val_num)
    
    remaining_elements = [file for file in image_filenames if file not in val_subset if file not in train_subset]

    
    #test 
    test_num = num_images - train_num - val_num
    test_subset = random.sample(remaining_elements,k=test_num)
    test_X.extend([os.path.join(test_dir,class_name,str(i)) for i in test_subset])
    test_Y.extend([class_name.split('.')[-1]]*test_num)
    
    remaining_elements = [file for file in image_filenames if file not in val_subset if file not in train_subset if file not in test_subset]


    # Create the destination directories for the current class
    os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(test_dir, class_name), exist_ok=True)
    
    

    # Move images to the training set folder
    for i in train_subset:
       src_path = os.path.join(class_dir, i)
       dst_path = os.path.join(train_dir, class_name, i)
       shutil.copy(src_path, dst_path)
    
    # Move images to the validation set folder
    for i in val_subset:
       src_path = os.path.join(class_dir, i)
       dst_path = os.path.join(val_dir, class_name, i)
       shutil.copy(src_path, dst_path)
    
    # Move images to the test set folder
    for i in test_subset:
       src_path = os.path.join(class_dir, i)
       dst_path = os.path.join(test_dir, class_name, i)
       shutil.copy(src_path, dst_path)
    
```

```python
print('Training set length is',len(train_X))
print('Validation set length is',len(val_X))
print('Test set length is',len(test_X))
val_X
```

```python
import glob,pandas as pd

train_X = list()
valid_X = list()
test_X = list()
train_Y = list()
valid_Y = list()
test_Y = list()
train_classes = list()
test_classes = list()
train_dir = "../Data/train/"
val_dir = "../Data/val/"
test_dir = "../Data/test/"

classes= ["baseball", "basketball", "fencing", "tennis", "volleyball"]

for folder in classes:
  # print(folder)
  files = glob.glob('../Data/Sport_Images/'+folder+'/*.jpg')
  
  num_train = math.ceil(0.7*len(files))
  temp_train = random.sample(files,k=num_train)
  train_X.extend(temp_train)
  train_Y.extend([folder.split('.')[-1]]*num_train)
  train_classes.extend([folder.split('.')[-1]]*num_train)

  num_valid = math.ceil(0.15*len(files))
  temp_valid = []
  for file in files:
    if file not in temp_train:
      temp_valid.append(file)
  temp_valid = random.sample(temp_valid, k = num_valid)
  valid_X.extend(temp_valid)
  valid_Y.extend([folder.split('.')[-1]]*num_valid)


  temp_test = []
  for file in files:
    if file not in temp_train:
      if file not in temp_valid:
        temp_test.append(file)
  test_X.extend(temp_test)
  test_Y.extend([folder.split('.')[-1]]*len(temp_test))
  test_classes.extend([folder.split('.')[-1]]*len(temp_test))

print(train_X[0],train_X[0].split('/')[-2],train_X[0].split('/')[-1])


for file in train_X:
       src_path = file
       dst_path = os.path.join(train_dir, file.split('/')[-2])
       os.makedirs(dst_path, exist_ok=True)  # create the folder if it doesn't exist
       shutil.copy(src_path, dst_path+'/'+file.split('/')[-1])

for file in test_X:
       src_path = file
       dst_path = os.path.join(val_dir, file.split('/')[-2], file.split('/')[-1])
       os.makedirs(dst_path, exist_ok=True)  # create the folder if it doesn't exist
       shutil.copy(src_path, dst_path)

for file in valid_X:
         src_path = file
         dst_path = os.path.join(test_dir, file.split('/')[-2], file.split('/')[-1])
         os.makedirs(dst_path, exist_ok=True)  # create the folder if it doesn't exist
         shutil.copy(src_path, dst_path)
  
#one hot encoding Y labels
train_Y = pd.get_dummies(train_Y).values
valid_Y = pd.get_dummies(valid_Y).values
test_Y = pd.get_dummies(test_Y).values



print('Training set length is',str(len(train_X)))
print('Validation set length is',str(len(valid_X)))
print('Test set length is',str(len(test_X)))
```
